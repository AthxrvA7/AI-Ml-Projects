```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from datasets import load_dataset
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
dataset = load_dataset("emotion")
df = pd.DataFrame(dataset['train'])
df.head()
```

```python
labels = ["sadness", "joy", "love", "anger", "fear", "surprise"]

df['emotions'] = df['label'].apply(lambda x: labels[x])
df.head()
```

```python
vectorizer = TfidfVectorizer()

x = vectorizer.fit_transform(df['text'])
y = df['label']

print(x.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)

model = LogisticRegression(max_iter=200)
model.fit(X_train, y_train)
```

```python
def predict_emotion(text):
    vec = vectorizer.transform([text])
    pred = model.predict(vec)[0]
    return labels[pred]

print(predict_emotion("I feel amazing"))
```

```python
def weighted_emotions(history):
    weights = {}
    n = len(history)

    for i, emotion in enumerate(history):
        weight = i + 1

        if emotion in weights:
            weights[emotion] += weight
        else:
            weights[emotion] = weight

    return weights
```

```python
def respond_to_emotion(emotion):
    if emotion == "joy":
        return "That's great to hear!"

    elif emotion == "sadness":
        return "I'm here for you."

    elif emotion == "anger":
        return "Take a deep breath."

    elif emotion == "fear":
        return "It's okay, you're safe."

    elif emotion == "surprise":
        return "That’s unexpected!"

    else:
        return "I understand."
```

```python
def emotional_agent(text):
    emotion = predict_emotion(text)

    emotion_history.append(emotion)

    if len(emotion_history) > 5:
        emotion_history.pop(0)

    weights = weighted_emotions(emotion_history)

    dominant = max(weights, key=weights.get)

    response = respond_to_emotion(dominant)

    return emotion, dominant, response
```

```python
while True:
    text = input("You: ")

    emotion, dominant, response = emotional_agent(text)

    print("Detected Emotion:", emotion)
    print("Mood:", dominant)
    print("Bot:", response)
```
